```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart_disease_data_with_features.csv')
df.head()
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

# Handling missing values
# Creating an imputer object for numerical columns with the strategy of using the median
imputer = SimpleImputer(strategy='median')

# Columns 'ca' and 'thal' are numerical and have missing values
df[['ca', 'thal']] = imputer.fit_transform(df[['ca', 'thal']])

# For categorical columns with missing values, we'll use the mode
# These columns are 'age_group', 'cholesterol_level', 'bp_level', and other columns with categorical data
cat_imputer = SimpleImputer(strategy='most_frequent')
df[['age_group', 'cholesterol_level', 'bp_level']] = cat_imputer.fit_transform(df[['age_group', 'cholesterol_level', 'bp_level']])

# Check if there are any more missing values
df.isnull().sum().sum()
```

```python
# Identify remaining missing values by column
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
# Imputing remaining missing values
df[['risk_factor', 'symptom_severity_diff']] = imputer.fit_transform(df[['risk_factor', 'symptom_severity_diff']])

df['avg_chol_by_age_group'] = imputer.fit_transform(df[['avg_chol_by_age_group']])

# Verify if all missing values are handled
df.isnull().sum().sum()
```

```python
df.isnull().sum()
```

```python
# Convert 'num' to a binary classification where 1-4 are mapped to 1, and 0 remains 0
df['num'] = df['num'].apply(lambda x: 1 if x > 0 else 0)

# Confirm the transformation by viewing the distribution of the target variable
df['num'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into features and the target variable
X = df.drop('num', axis=1)  # All other columns are features
y = df['num']  # Target variable

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Check the shape of the resulting sets
(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Initialize the StandardScaler
scaler = StandardScaler()

# Identifying categorical columns for one-hot encoding
categorical_features = df.select_dtypes(include=['object']).columns.tolist()

# Creating a column transformer with one-hot encoding for categorical features
column_transformer = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'  # apply passthrough for other columns not specified
)

# Apply the column transformer to training and test data
X_train_transformed = column_transformer.fit_transform(X_train)
X_test_transformed = column_transformer.transform(X_test)

# Apply scaling to the transformed data
X_train_scaled = scaler.fit_transform(X_train_transformed)
X_test_scaled = scaler.transform(X_test_transformed)

# Confirm scaling by checking the mean of the scaled training features (should be close to 0)
X_train_scaled.mean(axis=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Logistic Regression model
log_reg = LogisticRegression()

# Fit the model on the scaled training data
log_reg.fit(X_train_scaled, y_train)

# Predicting the test set results
y_pred = log_reg.predict(X_test_scaled)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(accuracy)
print(report)
```

```python
from sklearn.model_selection import GridSearchCV

# Parameters for grid search
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']  # Algorithm to use in optimization
}

# Create the GridSearchCV object with logistic regression
grid_search = GridSearchCV(estimator=LogisticRegression(max_iter=1000), param_grid=param_grid, cv=5, scoring='accuracy')

# Fit GridSearchCV
grid_search.fit(X_train_scaled, y_train)

# Best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

(best_params, best_score)
```

```python
from sklearn.model_selection import cross_val_score

# Create a new logistic regression model with the best parameters
optimized_log_reg = LogisticRegression(C=1, solver='lbfgs', max_iter=1000)

# Perform cross-validation
cv_scores = cross_val_score(optimized_log_reg, X_train_scaled, y_train, cv=5, scoring='accuracy')

# Calculate the mean and standard deviation of the cross-validated scores
cv_mean = cv_scores.mean()
cv_std = cv_scores.std()

(cv_mean, cv_std)
```

```python
# Train the optimized logistic regression model on the entire training set
optimized_log_reg.fit(X_train_scaled, y_train)

# Predict the test set results
y_test_pred = optimized_log_reg.predict(X_test_scaled)

# Calculate accuracy and other performance metrics on the test set
final_accuracy = accuracy_score(y_test, y_test_pred)
final_report = classification_report(y_test, y_test_pred)

print(final_accuracy)
print(final_report)
```
