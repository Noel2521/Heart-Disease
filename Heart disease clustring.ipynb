```python
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Preparing the data
X = data.drop('Cluster', axis=1).select_dtypes(include=[np.number])  # Selecting only numeric features
y = data['Cluster']  # Labels

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model Training
model = LogisticRegression(max_iter=1000, solver='lbfgs')
model.fit(X_train, y_train)

# Model Evaluation
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

accuracy, conf_matrix
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Creating a pipeline that includes scaling and the logistic regression model
pipeline = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))

# Fitting the model with the training data
pipeline.fit(X_train, y_train)

# Predicting the test data
y_pred = pipeline.predict(X_test)

# Evaluating the model with scaled features
scaled_accuracy = accuracy_score(y_test, y_pred)
scaled_conf_matrix = confusion_matrix(y_test, y_pred)

scaled_accuracy, scaled_conf_matrix
```

```python
import matplotlib.pyplot as plt

# Extracting coefficients from the logistic regression model within the pipeline
coefficients = pipeline.named_steps['logisticregression'].coef_

# Creating a DataFrame to hold the coefficients for each cluster
coef_df = pd.DataFrame(coefficients, columns=X_train.columns, index=[f"Cluster {i}" for i in range(coefficients.shape[0])])

# Plotting the coefficients for each cluster
fig, ax = plt.subplots(figsize=(14, 6))
coef_df.T.plot(kind='bar', ax=ax)
plt.title('Feature Importance by Cluster')
plt.ylabel('Coefficient Magnitude')
plt.axhline(0, color='black', linewidth=0.8)
plt.xticks(rotation=45)
plt.legend(title='Cluster')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Return the coefficient data frame for detailed review
coef_df
```

```python
# Selecting key features based on significant coefficients across clusters
key_features = coef_df.columns[(coef_df.abs() > 0.5).any()].tolist()

# Preparing data with only key features
X_key_features = X[key_features]

# Splitting the data into training and testing sets with key features
X_train_key, X_test_key, y_train, y_test = train_test_split(X_key_features, y, test_size=0.3, random_state=42)

# Creating a pipeline that includes scaling and logistic regression
key_feature_pipeline = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))

# Fitting the model with the training data using only key features
key_feature_pipeline.fit(X_train_key, y_train)

# Predicting the test data
y_pred_key = key_feature_pipeline.predict(X_test_key)

# Evaluating the model with scaled key features
key_feature_accuracy = accuracy_score(y_test, y_pred_key)
key_feature_conf_matrix = confusion_matrix(y_test, y_pred_key)

key_feature_accuracy, key_feature_conf_matrix
```
